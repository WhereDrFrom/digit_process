```python
# 三角波
import os
from thinkdsp import TriangleSignal
from thinkdsp import decorate
signal = TriangleSignal(200)
signal.plot()
```

```python
# 锯齿波
from Generator import SawtoothSignal
SawSignal = SawtoothSignal(200)
SawSignal.plot()
```

```python
#三角波的频谱
wave = signal.make_wave(duration=0.5, framerate=10000)
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
# 锯齿波的频谱
SawWave = SawSignal.make_wave(duration=0.5, framerate=10000)
SawSpectrum = SawWave.make_spectrum()
SawSpectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
#矩形波 
from thinkdsp import SquareSignal  

signal = SquareSignal(200)
duration = signal.period*3
segment = signal.make_wave(duration, framerate=10000)
segment.plot()
decorate(xlabel='Time (s)')
```

```python
# 矩形波的频谱
spectrum = segment.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
#矩形波1100Hz,采样10000的频谱
signal = SquareSignal(1100)
duration = signal.period*3
SqarSegment = signal.make_wave(duration, framerate=10000)
spec = SqarSegment.make_spectrum()
spec.plot()
decorate(xlabel='Frequency (Hz)')
```

```python
# 上方矩形波对应的音频
SqarSegment.apodize()
SqarSegment.make_audio()
```

```python
# 输出频谱的前几个值
for i in spec.fs:
    print(i,end=" ")
```

```python
# 创建440Hz的三角波波形
signal = TriangleSignal(440)
wave = signal.make_wave(duration=0.01, framerate=20000)
wave.plot()
```

```python
# 三角波的频谱
spectrum = wave.make_spectrum()
spectrum.plot()
decorate(xlabel='Frequency (Hz)')
```
